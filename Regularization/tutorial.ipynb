```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
dir(diabetes)
```

```python
df=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df
```

```python
df['Target']=diabetes.target
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
features=df[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']]
scaled_features=scaler.fit_transform(features)
df[features.columns]=scaled_features
df[['Target']]=scaler.fit_transform(df[['Target']])
df
```

```python
X=df.drop(['Target'],axis='columns')
y=df[['Target']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
lr=RandomizedSearchCV(LinearRegression(),{},cv=5,n_iter=10,return_train_score=False)

tree=RandomizedSearchCV(DecisionTreeRegressor(),{
    'max_leaf_nodes':[10,20,30,40],
    'criterion':['absolute_error', 'friedman_mse', 'squared_error', 'poisson']
},cv=5,n_iter=10,return_train_score=False)

rf=RandomizedSearchCV(RandomForestRegressor(),{
    'n_estimators':[1,5,10,20]
},cv=5,n_iter=10,return_train_score=False)

svm=RandomizedSearchCV(SVR(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear','poly','sigmoid']
},cv=5,n_iter=10,return_train_score=False)
```

```python
lr.fit(X,y)
tree.fit(X,y)
rf.fit(X,y)
svm.fit(X,y)
```

```python
print(f"Linear Regression: {lr.best_params_} \t Best Score: {lr.best_score_}")
print(f"Decision Tree: {tree.best_params_} \t Best Score: {tree.best_score_}")
print(f"Random Forest: {rf.best_params_} \t Best Score: {rf.best_score_}")
print(f"SVM: {svm.best_params_} \t Best Score: {svm.best_score_}")
```

As we can see none of the Regression models give an appropriate accuracy score, hence we may apply Regularization techniques like L1(Lasso), L2(Ridge) or Elastic Net to reduce overfitting.

#### **Lasso (L1)** - penalizes the features that have low predictive outcomes by shrinking their coeffiients to 0.
#### Penalty term = Î»âˆ‘|Î²â±¼| , where Î²â±¼ are the coefficients in the regression model

#### **Ridge (L2)** - penalizes features that have low predictive outcomes by shrinking their coefficients closer to 0.
#### Penalty term = Î»âˆ‘Î²â±¼**2 , where Î²â±¼ are the coefficients in the regression model

#### **Elastic Net** - weighted combination of both L1 and L2, combines both feature elimination from Lasso and feature coefficient reduction from Ridge
#### Penalty Term = Î»( âˆ‘|Î²â±¼| + âˆ‘Î²â±¼**2)

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
lasso=RandomizedSearchCV(Lasso(),{
    'alpha':[0.01,0.1,1,10,100]
},cv=4,n_iter=2,return_train_score=False)

ridge=RandomizedSearchCV(Ridge(solver='auto'),{
    'alpha':[0.01,0.1,1,10,100]
},cv=4,n_iter=2,return_train_score=False)

elastic_net=RandomizedSearchCV(ElasticNet(),{
    'alpha':[0.01,0.1,1,10,100],
    'l1_ratio':[0.1,0.5,0.7,0.9,1]
},cv=4,n_iter=2,return_train_score=False)
```

```python
lasso.fit(X,y)
ridge.fit(X,y)
elastic_net.fit(X,y)
```

```python
print(f'Lasso: {lasso.best_params_} \t Score: {lasso.best_score_}')
print(f'Ridge: {ridge.best_params_} \t Score: {ridge.best_score_}')
print(f'Elastic net: {elastic_net.best_params_} \t Score: {elastic_net.best_score_}')
```
