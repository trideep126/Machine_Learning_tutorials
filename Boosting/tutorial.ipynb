```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
dir(diabetes)
```

```python
df=pd.DataFrame(diabetes.data,columns=diabetes.feature_names)
df
```

```python
df['Target']=diabetes.target
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
features=df[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']]
scaled_features=scaler.fit_transform(features)
df[features.columns]=scaled_features
df[['Target']]=scaler.fit_transform(df[['Target']])
df
```

```python
X=df.drop(['Target'],axis='columns')
y=df[['Target']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
lr=RandomizedSearchCV(LinearRegression(),{},cv=5,n_iter=10,return_train_score=False)

tree=RandomizedSearchCV(DecisionTreeRegressor(),{
    'max_leaf_nodes':[10,20,30,40],
    'criterion':['absolute_error', 'friedman_mse', 'squared_error', 'poisson']
},cv=5,n_iter=10,return_train_score=False)

rf=RandomizedSearchCV(RandomForestRegressor(),{
    'n_estimators':[1,5,10,20]
},cv=5,n_iter=10,return_train_score=False)

svm=RandomizedSearchCV(SVR(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear','poly','sigmoid']
},cv=5,n_iter=10,return_train_score=False)
```

```python
lr.fit(X,y)
tree.fit(X,y)
rf.fit(X,y)
svm.fit(X,y)
```

```python
print(f"Linear Regression: {lr.best_params_} \t Best Score: {lr.best_score_}")
print(f"Decision Tree: {tree.best_params_} \t Best Score: {tree.best_score_}")
print(f"Random Forest: {rf.best_params_} \t Best Score: {rf.best_score_}")
print(f"SVM: {svm.best_params_} \t Best Score: {svm.best_score_}")
```

As we can see none of the Regression models give an appropriate accuracy score, hence we may apply some other Boosting Techniques

```python
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
```

#### **Adaboost**(Adaptive Boosting) - sequentially adds predictors to an ensemble, where each predictor focuses on the errors made by previous predictors.
Simple and less prone to overfitting but can be sensitive to noisy data and outliers

#### **Gradient Boosting** - sequentially builds an ensemble of trees where each new tree attempts to correct the errors of the combined ensemble of all previous trees by minimizing a loss function.
Provides high prediction accuracy but can be computationally intensive

#### **XGboost** - optimized version of gradient boosting, incorporates enhancements like tree pruning, regularization.
Highly efficient and scalable but requires careful tuning of hyperparameters for optimal performance

```python
ab= RandomizedSearchCV(AdaBoostRegressor(),{
    'n_estimators':[100,200,300],
    'learning_rate':[0.01,0.1,0.2,0.5],
    'loss':['linear', 'square', 'exponential']
},cv=4,n_iter=3,return_train_score=False)

gb=RandomizedSearchCV(GradientBoostingRegressor(),{
    'n_estimators':[100,200,300],
    'learning_rate':[0.01,0.1,0.2,0.5],
    'max_depth':[3,5,7],
    'loss':['squared_error', 'absolute_error', 'huber', 'quantile'],
    'subsample':[0.6,0.8,1.0],
    'criterion':['friedman_mse', 'squared_error']
},cv=4,n_iter=3,return_train_score=False)

xgb=RandomizedSearchCV(XGBRegressor(),{
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
},cv=4,n_iter=3,return_train_score=False)
```

```python
ab.fit(X,y)
gb.fit(X,y)
xgb.fit(X,y)
```

```python
print(f"AdaBoost: {ab.best_params_} \t Score: {ab.best_score_}")
print(f"Gradient Boosting: {gb.best_params_} \t Score: {gb.best_score_}")
print(f"XGBoost: {xgb.best_params_} \t Score: {xgb.best_score_}")
```
